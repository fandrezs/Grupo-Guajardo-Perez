```python
import altair as alt
import pandas as pd
import unicodedata

alt.data_transformers.disable_max_rows()

source = pd.read_csv('/content/Nueva BD (oficial) XBOX - FELIPE GUAJARDO.csv', sep=";", encoding='utf-8')

source.head()
```

```python
import altair as alt
import pandas as pd
import unicodedata

source = source [["Ano", "Videojuego", "Num de trabajadores", "Tamano (MB)"]]

print (source)
```

```python
source["Ano"] = pd.to_numeric(source["Ano"], errors="coerce")
```

```python
source_filtrado = source [(source["Ano"] >= 2000) & (source["Ano"] <= 2026)]
print (source_filtrado)
```

```python
resumen = source_filtrado.groupby("Ano").agg({
    "Num de trabajadores": "sum",
    "Videojuego": "count"
}).reset_index()

resumen.columns = ["Ano", "Total_trabajadores", "Cantidad_videojuegos"]

resumen.head()
```

```python
import altair as alt

line_trabajadores = alt.Chart(resumen).mark_line(
    point=True,
    color='#D93D86'
).encode(
    x=alt.X('Ano:O', title='AÃ±o'),
    y=alt.Y(
        'Total_trabajadores:Q',
        title='Cantidad de trabajadores',
        axis=alt.Axis(titleColor='#D93D86')
    ),
    tooltip=['Ano', 'Total_trabajadores']
)

line_videojuegos = alt.Chart(resumen).mark_line(
    point=True,
    color='#ECD14D'
).encode(
    x=alt.X('Ano:O', title='AÃ±o'),
    y=alt.Y(
        'Cantidad_videojuegos:Q',
        title='Cantidad de videojuegos lanzados',
        axis=alt.Axis(titleColor='#ECD14D')
    ),
    tooltip=['Ano', 'Cantidad_videojuegos']
)

chart_combined = alt.layer(
    line_trabajadores,
    line_videojuegos
).resolve_scale(
    y='independent'
).properties(
    title='Cantidad de trabajadores y videojuegos lanzados de Xbox Games Studios por aÃ±o',
    width=600,
    height=300
)

final_chart = chart_combined.configure(
    background='#000000'
).configure_view(
    fill='#000000'
).configure_axis(
    labelColor='#DDCEBB',
    titleColor='#DDCEBB'
).configure_title(
    color='#DDCEBB'
)

final_chart
```
