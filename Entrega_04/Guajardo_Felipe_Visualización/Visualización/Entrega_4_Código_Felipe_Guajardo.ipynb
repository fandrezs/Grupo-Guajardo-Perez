```python
import altair as alt
import pandas as pd
import unicodedata

alt.data_transformers.disable_max_rows()

source = pd.read_csv('/content/Nueva BD (oficial) XBOX - FELIPE GUAJARDO.csv', sep=";", encoding='utf-8')

source.head()
```

```python
import altair as alt
import pandas as pd
import unicodedata

source = source [["Ano", "Videojuego", "Num de trabajadores", "Tamano (MB)"]]

print (source)
```

```python
source["Ano"] = pd.to_numeric(source["Ano"], errors="coerce")
```

```python
source_filtrado = source [(source["Ano"] >= 2000) & (source["Ano"] <= 2026)]
print (source_filtrado)
```

```python
resumen = source_filtrado.groupby("Ano").agg({
    "Num de trabajadores": "sum",
    "Videojuego": "count"
}).reset_index()

resumen.columns = ["Ano", "Total_trabajadores", "Cantidad_videojuegos"]

resumen.head()
```

```python
import altair as alt

chart_trabajadores = alt.Chart(resumen).mark_line(
    point=True,
    color='pink'
).encode(
    x=alt.X('Ano:O', title='Año'),
    y=alt.Y('Total_trabajadores:Q', title='Suma de trabajadores'),
    tooltip=['Ano', 'Total_trabajadores']
).properties(
    title='Total de trabajadores por año',
    width=600,
    height=150
)

chart_trabajadores
```

```python
chart_trabajadores.save("chart.html")
```

```python
import altair as alt

chart_videojuegos = alt.Chart(resumen).mark_line(
    point=True,
    color='yellow'
).encode(
    x=alt.X('Ano:O', title='Año'),
    y=alt.Y('Cantidad_videojuegos:Q', title='Cantidad de videojuegos lanzados'),
    tooltip=['Ano', 'Cantidad_videojuegos']
).properties(
    title='Videojuegos lanzados por año entre el 2000 y 2026',
    width=600,
    height=150
)

chart_videojuegos
```

```python
chart_videojuegos.save("chart.html")
```
