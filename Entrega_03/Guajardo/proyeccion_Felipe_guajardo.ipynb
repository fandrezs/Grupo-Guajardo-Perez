```python
import altair as alt
import pandas as pd
import unicodedata

alt.data_transformers.disable_max_rows()

source = pd.read_csv('/content/BD Actualizada juegos Xbox - Felipe Guajardo.csv', sep=";", encoding='utf-8')

source.head()
```

```python
import altair as alt
import pandas as pd
import unicodedata

source_filtrado = source[source['Compania'] == 'The Coalition']
source_filtrado = source_filtrado[['Videojuego', 'Tamano (MB)', 'Num de trabajadores']]

print(source_filtrado)
```

```python
import altair as alt
alt.renderers.enable('colab')

chart = alt.Chart(source_filtrado).mark_circle(size=60).encode(
    x='Tamano (MB)',
    y='Num de trabajadores',
    tooltip=['Videojuego', 'Tamano (MB)', 'Num de trabajadores']
).properties(
    title='The Coalition: Tamaño del Juego vs Número de Trabajadores'
)

chart.show()
chart.save("chart.html")
```
